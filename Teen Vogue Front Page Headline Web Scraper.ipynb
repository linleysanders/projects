```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get('https://www.teenvogue.com')
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
stories = doc.find_all('h3', class_='feature-item-hed')
stories
```

```python
for story in stories:
    print(story.text)
```

```python
url_bank = []

for urls in doc.find_all('a', class_='feature-item-link feature-item-link-hed', href=True):
    print("http://www.teenvogue.com"+ urls['href'])
    if urls not in url_bank:
        url_bank.append(urls)
```

```python
for story in stories:
    print(story.text)
    print("----")
for urls in doc.find_all('a', class_='feature-item-link feature-item-link-hed', href=True):
    print("http://www.teenvogue.com" + urls['href'])
```

```python
list_of_stories = []

for story in stories:
    story_dict = {}
    story_dict['headline'] = story.text
    story_dict['URL'] = url_bank
    list_of_stories.append(story_dict)

if urls not in url_bank:
    url_bank.append(urls)
    
list_of_stories
```

```python
df = pd.DataFrame(list_of_stories)
df.head()
```
