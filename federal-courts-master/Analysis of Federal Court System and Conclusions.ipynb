```python
df = pd.read_csv("final_court_of_appeals_map.csv")
df.head()
```

```python
gender_table = df.groupby('properties.name')['properties.headline'].value_counts().reset_index(name = 'gender_counts')
gender_table.head()
```

```python
gender_table = df.groupby('properties.name')['properties.headline'].value_counts().unstack()
gender_table[:-1]
```

```python
dfd = pd.read_csv("final_district_courts_map.csv")
dfd.head()
```

```python
gender_table_dfd = dfd.groupby('properties.name')['properties.headline'].value_counts().unstack()
gender_table_dfd.head(10)
```

```python
df_gender_two = df.groupby('properties.name')['properties.headline'].value_counts().unstack().reset_index()
df_gender_two
```

```python
dfd_gender_two = dfd.groupby('properties.name')['properties.headline'].value_counts().unstack().reset_index()
dfd_gender_two.head()
```

# Even the most female-representative region for the Court of Appeals has 56 percent male judges

```python
df_gender_two['Total_Judges'] = df_gender_two['Male'] + df_gender_two['Female']
```

```python
df_gender_two['Percent_Male'] = df_gender_two['Male'] / df_gender_two['Total_Judges']
```

```python
df_gender_two.sort_values(by = 'Percent_Male', ascending=False)[:-1]
```

```python
gender_table_dfd['Total_Judges'] = gender_table_dfd['Male'] + gender_table_dfd['Female']
```

```python
gender_table_dfd['Percent_Male'] = gender_table_dfd['Male'] / gender_table_dfd['Total_Judges']
```

```python
gender_table_dfd.sort_values(by = 'Percent_Male', ascending=False)
```
